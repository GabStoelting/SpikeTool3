This script can be used to analyze the following parameters per condition:
* Mean calcium per condition
For usage, the <i>filename</i> variable must be changed accordingly. <br>
<br>
As is, the display only differentiates between angiotensin-II concentrations. Please change accordingly if other or more parameters shall be investigated.<br>
<br>

<i>Version: 20200503_1_GS</i>

```python
import calim
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import pandas as pd
```

```python
# Stand Kalibration 10.03.2020 für das Scientifica-Setup
# kd = 1990.212439237825
# R_max = 1.9267297757725788
# R_min = 0.27585714285714286
# sf2_by_sb2 = 1.0011635251408988

# Stand Kalibration 18.10.2019 für das Scientifica-Setup
kd = 1213.86029776
R_max = 1.789857623786418
R_min = 0.27880000000000005
sf2_by_sb2 = 1.0316526179375363

def calcium(data, kd, R_max, R_min, sf2_by_sb2):
    return kd*((data-R_min)/(R_max-data))*sf2_by_sb2
```

```python

filename = "Fura2AM_Angiotensin.pkl"

data_pkl = pickle.load(open(filename, "rb"))
```

```python
# The DataFrame "conditions" will contain all information about conditions
data = pd.DataFrame()
print(f"Processing recordings contained in {filename}:")

for recording in data_pkl.recordings:
    print(recording)
    
    r = data_pkl.recordings[recording]

    info = {desc: r.information[desc] for desc in r.information}
    
    for cell in r.cells:
        c = r.cells[cell]
        
        if c.use:
            for con in r.conditions:
                info["recording"] = recording
                info["start"] = int(con.start)
                info["end"] = int(con.end)
                info["cell"] = cell
                info["dt"] = r.dt # Add "dt" aka seconds/frame
                info.update({desc: con.information[desc] for desc in con.information}) # Add all further condition variables
                info["mean_raw"]  = c.raw_data[info["start"]:info["end"]].mean()
                info["mean_ca"] = calcium(c.raw_data[info["start"]:info["end"]], kd, R_max, R_min, sf2_by_sb2).mean()
                
                data = data.append(pd.Series(info), ignore_index=True)
print("Done.")
```

```python
x_cat = "angiotensin"
x_label = "[Angiotensin-II] (pM)"
y_cat = "mean_ca"
y_label = "Mean Ca$^{2+}$ (nM)"

plt.figure()
sns.catplot(x=x_cat, y=y_cat, data=data, kind="box")

# The following will plot the number of observations (aka "n").
# Please change this if a custom order of values is specified!
for i, value in enumerate(data[x_cat].unique()):
    plt.text(i, data[data[x_cat]==value][y_cat].max()+25, len(data[data[x_cat] == value]), 
            horizontalalignment="center", verticalalignment="bottom")


plt.title("Mean Ca$^{2+}$")
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.show()

# Plot per animal

plt.figure()
sns.catplot(x=x_cat, y=y_cat, hue="cell", col="recording", data=data, aspect=0.5)
plt.show()
```
