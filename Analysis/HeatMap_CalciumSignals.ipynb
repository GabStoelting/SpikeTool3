```python
def drawHeatMap(rec, title, pxPerCell = 50, norm=False, cond=False):
    
    output = np.array([])

    for cell in rec.cells:  
        c = r.cells[cell]

        if c.use:
            # Plot data either normed (only within each cell!) or not
            if norm:
                output = np.append(output, np.tile(c.raw_data/np.max(c.raw_data), (pxPerCell, 1)))
            else:
                output = np.append(output, np.tile(c.raw_data, (pxPerCell, 1)))

    output = output.reshape(-1, len(c.raw_data))
    print(output.shape)
    plt.figure()
    plt.matshow(output)
    
    # Plot red lines at the end of each condition segment if cond==True
    if cond:
        for condition in c.conditions:
            plt.axvline(condition.end, color="red")
    plt.title(title)
    #plt.tight_layout()
    plt.show()
```

```python
for rec in data_hdf.recordings:
    r = data_hdf.recordings[rec]
    drawHeatMap(r, rec, pxPerCell=100, norm=True, cond=True)
```
